```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
#Load the dataset
df=pd.read_csv('Bengaluru_House_Data.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Dropping all the unwanted columns
df=df.drop(["area_type","availability","society","balcony"],axis=1)
df.info()
```

```python
#Handling missing values in location
df['location'].value_counts()     # As whitefield is repeated for more times we fill it with the same
df['location']=df['location'].fillna('Whitefield')
df.info()
```

```python
#Handling missing values in size column
df['size'].value_counts()     # As 2BHK is repeated for more times we fill it with the same
df['size']=df['size'].fillna('2 BHK')
df.info()
```

```python
#Handling missing values in bath column
df['bath']=df['bath'].fillna(df['bath'].median())
df.info()
```

```python
df.head()
```

```python
#Cleaning size column
df['size'].unique()
out=[i.split()[0] for i in df['size']]
# out=[int(i.split()[0]) for i in df['size']]        # Another method to convert it into int datatype
df['bhk']=out
df.head()
```

```python
df.info()
```

```python
#Convert bhk column from object to int datatype
df['bhk']=df['bhk'].astype('int')
df.info()
```

```python
#Cleaning total_sqft
df['total_sqft'].unique()
```

```python
def convert(a):
    l=a.split('-')
    if len(l)==2:
        return float(l[0])+float(l[1])/2
    try:
        return float(a)
    except:
        return None
    
df['total_sqft']= df['total_sqft'].apply(convert)
df['total_sqft']= df['total_sqft'].fillna(df['total_sqft'].median())
df.info()
```

```python
df['price_per_sqft']= df['price']*100000/df['total_sqft']
df['price_per_sqft']
```

```python
df.head()
```

```python
#Cleaning location column
df['location']=df['location'].apply(lambda x:x.strip())
df['location'].value_counts()
```

```python
location_count=df['location'].value_counts()
loc_less_than_10=location_count[location_count<=10]
loc_less_than_10
```

```python
df['location']=df['location'].apply(lambda x:'Other' if x in loc_less_than_10 else x)
df['location'].value_counts()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.head()
```

```python
#Handling outlier in total_sqft
df=df[(df['total_sqft']/df['bhk'])>=300]
df.describe()
```

```python
#Handling outlier in total_sqft                                (Another method)
df=df[df['total_sqft']>= df['bhk']*300]
df.describe()
```

```python
#Handling outlier in bath
sns.boxplot(y='bath',data=df)                #Outliers
```

```python
df=df[df['bath']<=df['bhk']]
sns.boxplot(y='bath',data=df) 
```

```python
#Handling oulier in bhk
df=df[df['bhk']<=6]
sns.boxplot(y='bhk',data=df)
```

```python
#Handling oulier in price_per_sqft
sns.boxplot(y='price_per_sqft',data=df)
```

```python
#handling outliers in price_per_sqft
q1= df['price_per_sqft'].quantile(0.25)
q3= df['price_per_sqft'].quantile(0.75)
IQR= q3-q1
low_bound= q1-0.5*IQR
up_bound= q3+0.5*IQR
df= df[(df['price_per_sqft']>=low_bound) & (df['price_per_sqft']<=up_bound)]
sns.boxplot(y='price_per_sqft',data=df)
```

```python
df.head()
```

```python
df=df.drop(['size','price_per_sqft'],axis=1)
df.head()
```

```python
df.reset_index(inplace=True)
df=df.drop('index',axis=1)
```

```python
copied_dataset=df.copy()
from sklearn.preprocessing import LabelEncoder
obj=LabelEncoder()
df['location']=obj.fit_transform(df['location'])
df.head()
```

```python
copied_dataset
```

```python
X=df.drop('price',axis=1)
y=df['price']
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=RandomForestRegressor(random_state=42)
params={
    "n_estimators":[100,150,200,250,300],
    "max_depth":[3,4,5,6]
}
```

```python
grid=GridSearchCV(estimator=model,param_grid=params,cv=5)
grid.fit(Xtrain,ytrain)
```

```python
print('training efficiency:',grid.score(Xtrain,ytrain))
print('testing efficiency:',grid.score(Xtest,ytest))
```

```python
with open('RFmodel.pkl','wb') as file:
    pickle.dump(grid,file)
```
